```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
vid = cv2.VideoCapture("1.mp4")
vid.set(1, 30*25*60)
```

```python
img_t = cv2.split(vid.read()[1])[0]
```

```python
vid.set(1, 30*25*60)
```

```python
vid.set(1, 30*25*60-30)
img_tau = cv2.split(vid.read()[1])[0]
```

```python
vid.set(1, 30*25*60-60)
img_tau = cv2.split(vid.read()[1])[0]
```

```python
mser = cv2.MSER_create()
regions, _ = mser.detectRegions(img_t)
```

```python
hull = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions]
```

```python
img = img_t.copy()
cv2.polylines(img, hull, 1, (0, 255, 0))
plt.imshow(img)
```

```python
cv2.imwrite("img1.png", img)
```

```python
regions[0]
```

```python
mask = np.zeros((img.shape[0], img.shape[1], 1), dtype=np.uint8)
for contour in hull:
    cv2.drawContours(mask, [contour], -1, (255, 255, 255), -1)
```

```python
text_only = cv2.bitwise_and(img, img, mask=mask)
```

```python
plt.imshow(text_only)
```

```python
cv2.imwrite("im2.png", text_only)
```

```python
regions, _ = mser.detectRegions(img_tau)
hull = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions]
img = img_tau.copy()
cv2.polylines(img, hull, 1, (0, 255, 0))
plt.imshow(img)
```

```python
mask = np.zeros((img.shape[0], img.shape[1], 1), dtype=np.uint8)
for contour in hull:
    cv2.drawContours(mask, [contour], -1, (255, 255, 255), -1)
```

```python
text_only = cv2.bitwise_and(img, img, mask=mask)
```

```python
plt.imshow(text_only)
```

```python
cv2.imwrite("im3.png", text_only)
```
