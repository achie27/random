```python
v = cv2.VideoCapture("1.mp4")
pfr = v.read()[1]
pfr = cv2.resize(pfr, (352, 288))
pfr = cv2.cvtColor(pfr, cv2.COLOR_BGR2HSV)
```

```python
def func(f1, f2):
    n, m = f1.shape[0], f1.shape[1]
    dif = (f1 - f2)**2
    
    return dif.sum()/m/n

fl, p = [], 0

while True:
    fr = v.read()
    if not fr[0]: break
    fr = fr[1]
    fr = cv2.resize(fr, (352, 288))
    fr = cv2.cvtColor(fr, cv2.COLOR_BGR2HSV)
    fl.append(func(pfr, fr))
    pfr = fr.copy()
    p+=1
```

```python
counter = 0
fd = []
hist = []
v = cv2.VideoCapture("1.mp4")

while v.isOpened():
    suc, fr = v.read()
    if not suc : break

    fr = cv2.resize(fr, (320, 240))
    gray_fr = cv2.cvtColor(fr, cv2.COLOR_BGR2GRAY)

    hist_fr = cv2.calcHist(
        [gray_fr], [0], None, [256], [0, 256]
    )

    if counter >= 10:
        tmp = np.abs(
            hist_fr-hist[counter-10]
        )
        f = np.sum(tmp)/(2*320*240)
        fd.append(f)

    # save the histogram
    hist.append(hist_fr)
    counter+=1
```

```python
shots = []
diff_fd = np.diff(fd)

for i in range(2, len(fd)-2):

    local_max=diff_fd[i+1] <= 0 and diff_fd[i] >= 0

    if fd[i] > 0.35 and local_max :
        shots.append(i - 10)
```
